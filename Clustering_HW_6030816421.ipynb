```python
# นำข้อมูลวันและเวลามาต่อกัน
Fire['ACQ_TIME']=Fire['ACQ_TIME']/100
Fire['ACQ_TIME']=Fire['ACQ_TIME'].astype(str)
Fire['ACQ_TIME']
```

```python
Fire['ACQ_TIME']=Fire['ACQ_TIME'].str.replace('.',':',regex=True)
Fire['DATE&TIME']=Fire['ACQ_DATE'].astype(str)+' '+Fire['ACQ_TIME']
Fire['DATE&TIME']=pd.to_datetime(Fire['DATE&TIME'])
Fire['DATE&TIME']
```

```python
Fire[Fire['DATE&TIME']<=(Fire['DATE&TIME'].iloc[0]+pd.Timedelta(days=7))]
```

```python
Fire['DT']=Fire['DATE&TIME']-Fire['DATE&TIME'][0]
Fire['DD']=(Fire['DT'].dt.total_seconds())/(60*60*24)
Fire
```

```python
Fire_7=Fire[Fire['DATE&TIME']<=(Fire['DATE&TIME'].iloc[0]+pd.Timedelta(days=7))]
Fire_7
```

```python
Fire_7 = Fire[["LATITUDE","LONGITUDE","DD"]].head(813)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
axe = Axes3D(fig)
axe.scatter(Fire_7['LATITUDE'], Fire_7['LONGITUDE'], Fire_7['DD'])
axe.set_xlabel('lat(X)')
axe.set_ylabel('lon(Y)')
axe.set_zlabel('day(Z)')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
```

```python
#สร้างตัวแปรงใหม่ขึ้นมาเพื่อเก็บค่า latitude ใน 7 วัน
latitude7 = np.array(Fire_7.values.tolist())
latitude7
```

```python
#สร้าง model ในการจำแนกข้อมูลขึ้นมา ณ ที่นี้ใช้ DBSCAN
model = DBSCAN(eps=1, min_samples=24)
model.fit_predict(Fire_7)
pred = model.fit_predict(Fire_7)

fig = plt.figure()
axes = Axes3D(fig)
axes.scatter(latitude7[:,0], latitude7[:,1], latitude7[:,2], c=model.labels_, s=300)
axes.view_init(azim=200)
plt.show()
```

```python
model_label = model.labels_
```

```python
model_label
```

```python
#ประมวลผลค่า silhouette 
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_score(latitude7, model_label)
```
